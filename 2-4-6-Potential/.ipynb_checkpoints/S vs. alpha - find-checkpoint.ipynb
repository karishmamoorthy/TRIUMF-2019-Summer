For 2-4-6 Potential

```python
import numpy as np
```

```python
alpha_vals = np.linspace(0.500001, 2.0/3.0, 50)
```

```python
import Var_Hex_Pot_Act_3
```

```python
s_vals = []
r_vals = []
p_vals = []
r10_vals = []

print("#    S     R    Phi_0,   r10,   case")
for a in alpha_vals:
    s, r, phi_0, case, r10 = Var_Hex_Pot_Act_3.get_S(a, D=4)
    s_vals.append(s)
    r_vals.append(r)
    p_vals.append(phi_0)
    r10_vals.append(r10)
    print(len(s_vals), s_vals[-1], r_vals[-1], p_vals[-1], r10_vals[-1], case)
```

```python
import matplotlib.pyplot as plt
%matplotlib nbagg
```

```python
plt.close()
plt.plot(alpha_vals[2:], s_vals[2:], 'r.-')
plt.grid(True)
plt.title("S vs. alpha, 2-4-6- potential")
```

```python
plt.close()
```

```python
plt.close()
plt.plot(alpha_vals[2:], r_vals[2:], 'b.-')
plt.grid(True)
plt.title("R vs. alpha, 2-4-6- potential")
```

```python
plt.close()
```

```python
plt.close()
plt.plot(alpha_vals, p_vals, 'g.-')
plt.title("Initial values vs. alpha")
plt.xlabel("alpha")
plt.ylabel("Phi_0")
plt.grid(True)
```

```python
plt.close()
```

```python
file = open("var_hex_pot_act_S_vals_3_dof4.txt", "w+")
for i in s_vals:
    file.write(str(i) + " \n")
file.close()

file = open("var_hex_pot_act_R_vals_3_dof4.txt", "w+")
for i in r_vals:
    file.write(str(i) + " \n")
file.close()

file = open("var_hex_pot_act_R10_vals_3_dof4.txt", "w+")
for i in r10_vals:
    file.write(str(i) + " \n")
file.close()

file = open("var_hex_pot_act_P_vals_3_dof4.txt", "w+")
for i in p_vals:
    file.write(str(i) + " \n")
file.close()
```
