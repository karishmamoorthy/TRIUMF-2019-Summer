# S vs. T

On attempting to solve for a scalar field, with the potential (in Minkowski space i.e. real space), $V_{Minkowski}(\phi)$, where $\phi$ is the function for the scalar field
.

A Wick Rotation was performed, taking the problem to the Euclidean space:

$V_{Euclidean}(\phi) = - V_{Minkowski}(\phi)$

And so, as adapted from a paper by Sidney Coleman in 1977, and my supervisors' work, the equation to solve becomes:

### $ \frac{d^{2}\phi}{d\rho^{2}} + \frac{(D - 1)}{\rho}\frac{d\phi}{d\rho} + \frac{dV_{Euclidean}}{d\phi} = 0 $

where the fundamental assumptions are:

a) Scalar field is only a function of spatial dimensions and is radially symmetric -> $\phi(\rho)$ (this is the fundamental ansatz behind the bubble/bounce approximation, since we are assuming a phase change)

**b) Boundary conditions are $\frac{d\phi}{d\rho} |_{\rho = 0}$ = 0;  $\phi(\rho = \infty)$ = 0**

We attempt to solve this for D = 3 and D = 4.

### --------------------------------------------------------------------------------------------------------------------

Once we find $\phi(\rho)$, we calculate the Euclidean action:

For D = 3:

$S_{Euclidean} = \int [(\frac{d\phi}{d\rho})^{2} - V_{Euclidean}(\phi)]d^{3}x $

   = $ \int\int\int [(\frac{d\phi}{d\rho})^{2} - V_{Euclidean}(\phi)]\rho^{2}sin(\theta) d\rho d\theta d\psi $

### $\Rightarrow S_{Euclidean}[\phi(\rho)] = 4\pi \int [(\frac{d\phi}{d\rho})^{2} - V_{Euclidean}(\phi(\rho))]\rho^{2} d\rho $

.

Similarly, for D = 4:

### $\Rightarrow S_{Euclidean}[\phi(\rho)] = 4\pi^{2} \int [(\frac{d\phi}{d\rho})^{2} - V_{Euclidean}(\phi(\rho))]\rho^{3} d\rho $

.

NOTE: Due to computational limitations, the thin wall approximation was used to calculate the action, at times. This is discussed further in other places.

This action is used to calculate the probability of the field existing in a state of true vacuum, for a given potential.

### --------------------------------------------------------------------------------------------------------------------

Here, we are analyzing the tunneling rate of some field during its first order phase transition in the early Universe.

So, we have:

$V_{Minkowski}(\phi) = \mu^{2} \phi^{2} - A \phi^{3} + \lambda \phi^{4}$

                                                                                                           --- Eqn. 1

Now, this potential is temperature dependent. In fact, it depends on temperature, as given by:

$V_{Minkowski}(\phi) = (\mu^{2} + T^{2})\phi^{2} - A \phi^{3} + \lambda \phi^{4}$

                                                                                                           --- Eqn. 2
                                                                                                          
where, T is the temperature, and Eqn. 1 is the T = 0 potential i.e. the (potential of the) theory we are working with.

For the sake of convenience, we redefine $\mu$ as $\mu_{eff}$, to give:

$V_{Minkowski}(\phi) = (\mu_{eff}(T))^{2} \phi^{2} - A \phi^{3} + \lambda \phi^{4}$

                                                                                                           --- Eqn. 3

where, $\mu_{eff}$ = $\sqrt{\mu^2 + T^2}$

.

And so, the goal is to calculate action $S_{Euclidean}$ (and consequently, tunneling rate) as a function of temperature $T$ 

### --------------------------------------------------------------------------------------------------------------------

Any Quartic 2-3-4 polynomial can be expressed as:

$V_{Minkowski}(\phi) = \Lambda^{4}.((\frac{3 - 4\alpha}{2}).(\frac{\phi}{v})^{2} - (\frac{\phi}{v})^{3}  + \alpha.(\frac{\phi}{v})^{4})$

                                                                                                           --- Eqn. 4

Comparing with the quartic potential of the scalar field in question:

$V_{Minkowski}(\phi) = (\mu_{eff}(T))^{2} \phi^{2} - A \phi^{3} + \lambda \phi^{4}$

we have:

$v$ = $\frac{3A + \sqrt(9 A^{2} - 32 \mu_{eff}^{2} \lambda)}{8 \lambda}$

$\alpha$ = $\frac {\lambda v}{A}$

$\Lambda^{4}$ = A $v^{3}$

### ----------------------------------------------------------------------------------------------------------------------

All the quantities above have dimensions. 

In the typically used convention $\hbar$ = c = $k_{B}$ = 1, 

[$\phi$] = E, 
[$\rho$] = E$^{-1}$ **,

[$\mu$] = E,
[A] = E,
[$\lambda$] = 1,

[$\Lambda$] = E,
[$v$] = E,
[$\alpha$] = 1,

[$V$] = E$^{4}$

[$T$] = E

and 
[$S$] = E (when d.o.f. = 3) and 1 (when d.o.f. = 4)

where E holds the dimensions of energy.

### ------------------------------------------------------------------------------------------------------------------------

** d.o.f matters (but not for the dimensions of $\rho$):

d.o.f. = 3, implies that here, we assume that Temperature acts as a scale of Time; $\rho$ = $\sqrt{x^2 + y^2 + z^2}$

Tunneling Probability $\sim$ $e^{-S_{Euclidean}/T}$

.

d.o.f. = 4, includes Time in $\rho$; $\rho$ = $\sqrt{x^2 + y^2 + z^2 + t^2}$ (c = 1); 

Tunneling Probability $\sim$ $e^{-S_{Euclidean}}$

### -------------------------------------------------------------------------------------------------------------------------

Now, to get a system that does not have it's basic design (like, for example, the location of "infinity", or the basic shape of the curve) depending on these parameters, we must use dimensionless quantities.

We can use a transformation that *conveniently* makes everything dimensionless (to solve the equation), and then allows us to return the dimensions to the problem (so that solution is not affected by energy scale of problem).

So, we do the following transformations:

$\phi$ = $\frac{\phi_{dimensionful}}{v}$    $\Leftrightarrow$   Setting $v$ = 1, in Eqn. 4

$V$ =  $\frac{V_{dimensionful}}{\Lambda^{4}}$    $\Leftrightarrow$   Setting $\Lambda$ = 1, in Eqn. 4

This makes things way more convenient.

Looking at Eqn. 2, we note that $\mu$ and $T$ have the same dimensions. We also note that, to maintain the above transformations, $T$ (and also $\mu$) must transform as:

$T$ = $T_{dimensionful} \frac{v}{\Lambda^{2}}$

Now, we know that:

### $S_{Euclidean}[\phi(\rho)] = k \int [(\frac{d\phi}{d\rho})^{2} - V_{Euclidean}(\phi(\rho))]\rho^{D - 1} d\rho $

We want a transformation: $S_{dimensionful}$ $\rightarrow$ $S$, such that:

$S$ = (scalar).$S_{dimensionful}$

So, we need to pull out the same factor from both $(\frac{d\phi}{d\rho})^{2}$ and $V_{Euclidean}(\phi(\rho))$. Since we already know the transformations for $V$ and $\phi$, we find the corresponding transformation for $\rho$:

$\rho$ = $\rho_{dimensionful} \frac{\Lambda^{2}}{v}$

And therefore, 

$S$ = $\frac{1}{\Lambda^{4}} (\frac{\Lambda^{2}}{v})^{D} S_{dimensionful}$

So, 

for D = 3,.....$S$ = $\frac{\Lambda^{2}}{v^3} S_{dimensionful}$ 

and for D = 4,.....$S$ = $\frac{\Lambda^{4}}{v^4} S_{dimensionful}$
