```python
output1 = Act_Temp_quick.theory(0.57, 1000, 1000, dof=3)
```

```python
output2 = Act_Temp_quick.theory(0.57, 1000, 1000, dof=4)
```

```python
t1_array = output1[0][1:-1] #got rid of T = 0 value because S/T is undefined there. Got rid of T_start value as S -> inf
s1_array = output1[1][1:-1]
#r1_array = output1[2][1:-1] #this part doesn't exist, like this, in the present version of Act_Temp_quick

st1_array = s1_array/t1_array
```

```python
t2_array = output2[0][:-1]
s2_array = output2[1][:-1]
#r2_array = output2[2][:-1] #this part doesn't exist, like this, in the present version of Act_Temp_quick
```

```python
plt.close()
plt.plot(t1_array, st1_array, 'r.-')
plt.title("S/T vs. T; D = 3")
plt.xlabel("Temperature")
plt.ylabel("S/T")
plt.ylim(-1000, 30000)
plt.xlim(output1[0][-1] + 10, output1[0][0] - 10)
plt.grid(True)
```

```python
plt.close()
```

```python
plt.close()
```

```python
plt.close()
plt.semilogy(t1_array, s1_array, 'r.-')
plt.title("S (log scale) vs. T; D = 3")
plt.xlabel("Temperature")
plt.ylabel("S")
#plt.ylim(-1000, 30000)
plt.xlim(output1[0][-1] + 10, output1[0][0] - 10)
plt.grid(True)
```

```python
plt.close()
```

```python
plt.close()
plt.plot(t2_array, s2_array, 'r.-')
plt.title("S vs. T; D = 4")
plt.xlabel("Temperature")
plt.ylabel("S")
plt.ylim(-1*1e5, 3*1e6)
plt.xlim(output2[0][-1] + 10, output2[0][0] - 10)
plt.grid(True)
```

This was plotted to compare with "S vs. T - brute force method". The new version of Act_Temp_quick doesn't return this. DO NOT RUN THE BELOW AGAIN.

```python
phi_array = 1000*np.linspace(0, 1.2, 1000)

n = len(output1[0]) - 1

V_contour_1 = lambda x: output1[-1][0][0]*x**2 - output1[-1][1]*x**3 + output1[-1][2]*x**4
V_contour_2 = lambda x: output1[-1][0][int(n/5)]*x**2 - output1[-1][1]*x**3 + output1[-1][2]*x**4
V_contour_3 = lambda x: output1[-1][0][int(2*n/5)]*x**2 - output1[-1][1]*x**3 + output1[-1][2]*x**4
V_contour_4 = lambda x: output1[-1][0][int(3*n/5)]*x**2 - output1[-1][1]*x**3 + output1[-1][2]*x**4
V_contour_5 = lambda x: output1[-1][0][int(4*n/5)]*x**2 - output1[-1][1]*x**3 + output1[-1][2]*x**4
V_contour_6 = lambda x: output1[-1][0][n]*x**2 - output1[-1][1]*x**3 + output1[-1][2]*x**4

plt.close()
plt.plot(phi_array, V_contour_1(phi_array), 'k-', label= "T = " + str(round(output1[0][0], 3)))
plt.plot(phi_array, V_contour_2(phi_array), 'm-', label= "T = " + str(round(output1[0][int(n/5)], 3)))
plt.plot(phi_array, V_contour_3(phi_array), 'b-', label= "T = " + str(round(output1[0][int(2*n/5)], 3)))
plt.plot(phi_array, V_contour_4(phi_array), 'g-', label= "T = " + str(round(output1[0][int(3*n/5)], 3)))
plt.plot(phi_array, V_contour_5(phi_array), 'y-', label= "T = " + str(round(output1[0][int(4*n/5)], 3)))
plt.plot(phi_array, V_contour_6(phi_array), 'r-', label= "T = " + str(round(output1[0][n], 3)))
plt.plot([0, 1000*1.2], [0, 0], 'k-', linewidth=2)
plt.title("V(Phi) at different temperatures")
plt.xlabel("Phi")
plt.ylabel("Potential")
plt.legend()
plt.grid(True)
```

```python
plt.close()
```
