# Compaing different approximations in caluclating the Euclidean action, in the Bounce/Bubble approximation

Generalizing "The Bounce Approximation for a general static potential" for a general D,

### $ \frac{d^{2}\phi}{d\rho^{2}} + \frac{(D - 1)}{\rho}\frac{d\phi}{d\rho} + \frac{dV_{Euclidean}}{d\phi} = 0 $

where the fundamental assumptions are:

a) $\rho = \sqrt{x_{1}^2 + x_{2}^2 +.....  x_{D}^2}$ so d.o.f. = D (all in Euclidean space)

b) Scalar field is only a function of spatial dimensions and is radially symmetric -> $\phi(\rho)$ (this is the fundamental ansatz behind the bubble/bounce approximation, since we are assuming a phase change)

##### c) Boundary conditions are $\frac{d\phi}{d\rho} |_{\rho = 0}$ = 0;  $\phi(\rho = \infty)$ = 0

and so,

### $S_{Euclidean}[\phi(\rho)] = k \int [\frac{1}{2}(\frac{d\phi}{d\rho})^{2} - V_{Euclidean}(\phi(\rho))]\rho^{D - 1} d\rho $

where for $D = 3, k = 4 \pi$ 

and for $D = 4, k = 4 \pi^{2}$

.
---
### Thin Wall Approximation (Epsilon approx.)

In the thin wall approximation, we get Euclidean action:

## $S_{Euclidean}[\phi(\rho)] =  k R^{D - 1} S_{1}  + \frac{k R^{D} \epsilon}{D}$

where, $\epsilon$ is the depth of the true vacuum potential well (in Minkowski space) w.r.t. false vacuum. For this approximation, $\epsilon$ is considered small (and is negative).

and

## $R = \frac{(D - 1).S_{1}}{-\epsilon}$ 
with
## $S_{1} = \int_{0}^{\phi_0} [\sqrt{2V_{Minkowski}(\phi)}] d\phi$

[Refer notes]


.
---
### Continuity Approximation (Version 3):

Since V($\phi$) = $\epsilon$ implies constant excess potential, and that is usually not the case, I have used a variation of the above expression:

## $S_{Euclidean}[\phi(\rho)] = k R^{D - 1} \int_{0}^{\phi_R} [\sqrt{2V_{Minkowski}(\phi)}] d\phi$     +  

## $k \int_{0}^{R} [\frac{1}{2}(\frac{d\phi}{d\rho})^{2} - V_{Euclidean}(\phi(\rho))]\rho^{D - 1} d\rho$


where $\phi_R$ = $\phi(\rho = R)$

and R = Radius of bubble, such that $\phi(\rho = R) \approx$ 10% of $\phi_0$


This stems from the fact that my code inevitably "undershoots" when peaks are almost the same height (i.e. thin wall approximation). So a solution *is found* when 1st and 2nd peaks are almost the same height... it just doesn't quite reach the 2nd peak's top. So instead of resolving from scratch and trying to approximate it analytically like Coleman did, we only use the symmetric well approximation for the region where the solution couldn't reach. The symmetric well approximation works perfectly in that region, since $\phi_{R}$ > 0, AND $\phi_{R}$ < $\phi_{localzero}$.


Note: $\phi_{R}$ -> 0 is logically equivalent to no longer needing this approximation. This occurs when $\phi_{0}$ is considerably less than $\phi_{localmax}$ 

and so...

**This approximation guarantess continuity with the region where the solution satisfies boundary conditions. BUT, since the approximation relies on the solution, which is by definition, one that doesn't satisfy the boundary conditions... the accuracy of $\phi(\rho)$ is brought into question.** 

**The biggest problem with my solution is that, $\phi(\rho)$ vs. $\rho$ does not steepen quickly enough as $\alpha$ decreases. This problem is especially evident as $\alpha$ reduces past 0.513 (for D = 3) and 0.52 (for D = 4). This is illustrated in the (r10 - R) vs. $\alpha$ graphs shown below. 
Hence, the epsilon approximation is used, alongside with the continuity approximation, in the Final Version (and Version 1).**


.
---
### Peak approximation:
##### when used in conjuction with the thin wall approximation, this is Version 2

The problem with the thin wall approximation, is that V($\phi$) = $\epsilon$ implies constant excess potential, and that is usually not the case. So, instead of using the $\epsilon$ term, we use:

## $S_{Euclidean}[\phi(\rho)] = k R^{D - 1} \int_{0}^{\phi_0} [\sqrt{2V_{Minkowski}(\phi)}] d\phi$     +  

## $k \int_{0}^{\rho_0} [\frac{1}{2}(\frac{d\phi}{d\rho})^{2} - V_{Euclidean}(\phi(\rho))]\rho^{D - 1} d\rho$

where $\phi(\rho = \rho_0) = \phi_0$


This seems like the more intuitive option, when compared to the continuity approximation... however, there is a MAJOR problem:

**it relies VERY heavily on the region of the solution that is most dubious. While the continuity approximation connects to the proper solution (i.e. when $\alpha$ is reasonable large), this peak connects to the thin wall approximation (i.e. when $\alpha$ is very small). So, the solution found is most problematic here... and hence the peak approximation cannot be used.**

Version 3 -> Continuity approximation

Version 2 -> Peak approximation alongside with epsilon approximation

Version 1 -> Continuity approximation alongside with epsilon approximation

Pure epsilon approximation -> Just the thin wall approximation

Final version -> Epsilon approximation used upto $\alpha < 0.507$, the continuity approximation used to where it is valid, and linearization between those 2 regions.

Note that, in Version 2, $S =$ 'None' was used to replace values which were not defined (because of negative values under a square-root)

Observe that Version 2, 3 and Pure epsilon approximation are discontinuous with the proper solution... and Version 1 has a discontinuity where epsilon approximation connects to the continuity approximation. Hence the need for linearization in the final version.
